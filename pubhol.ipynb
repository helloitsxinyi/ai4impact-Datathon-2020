```python
import pandas as pd
import numpy as np
from datetime import datetime
%matplotlib inline

# read data and store in dataframes
raw_data = pd.read_csv('Raw_Data\data.csv')

raw_data['datetime'] = pd.to_datetime(raw_data['datetime'], format = "%d/%m/%Y %H:%M",)
#raw_data.drop(["Unnamed: 0"], axis = 1, inplace = True)
raw_data.set_index('datetime', inplace=True)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as plticker
plt.rcParams["figure.figsize"] = [50, 30]
plt.rcParams.update({'font.size': 22})
```

```python
### EASTER MONDAY 6 APR ###

## 2014 = orangered, 2015 = seagreen, 2016 = deeppink ##

## power data available for 2014, 2015, 2016:
# PH 2014: "2014-08-15", "2014-11-01", "2014-12-08", "2014-12-25", "2014-12-26"
# PH 2015: "2015-01-01", "2015-01-06" (Epiphany), "2015-04-06" (Easter Monday), "2015-04-25" (Liberation Day), "2015-05-01" (Labour Day), "2015-06-02" (Republic Day), "2015-08-15" (Assumption Day), "2015-11-01" (All Saints Day), "2015-12-08" (Immaculate Conception), "2015-12-25" (Christmas Day), "2015-12-26" (Saint Stephens Day)
# PH 2016: "2016-01-01", "2016-01-06", "2016-03-28" (Easter Monday), "2016-04-25", "2016-05-01"

lab15 = raw_data.loc["2015-05-01"]
lab15["Hour"] = raw_data["2015-05-01"].index.to_series().apply(lambda x: x.strftime("%H:%M"))
lab16 = raw_data.loc["2016-05-01"]
lab16["Hour"] = raw_data["2016-05-01"].index.to_series().apply(lambda x: x.strftime("%H:%M"))

rep15 = raw_data.loc["2015-06-02"]
rep15["Hour"] = raw_data["2015-06-02"].index.to_series().apply(lambda x: x.strftime("%H:%M"))

ad14 = raw_data.loc["2014-08-15"]
ad14["Hour"] = raw_data["2014-08-15"].index.to_series().apply(lambda x: x.strftime("%H:%M"))
ad15 = raw_data.loc["2015-08-15"]
ad15["Hour"] = raw_data["2015-08-15"].index.to_series().apply(lambda x: x.strftime("%H:%M"))

asd14 = raw_data.loc["2014-11-01"]
asd14["Hour"] = raw_data["2014-11-01"].index.to_series().apply(lambda x: x.strftime("%H:%M"))
asd15 = raw_data.loc["2015-11-01"]
asd15["Hour"] = raw_data["2015-11-01"].index.to_series().apply(lambda x: x.strftime("%H:%M"))

label = ["2015", "2016"]

fig, ax = plt.subplots(2,2)

ax[0,0].scatter(lab15.Hour, lab15["Power"], color = "seagreen")
ax[0,0].scatter(lab16.Hour, lab16["Power"], color = "deeppink")

ax[0,1].scatter(rep15.Hour, rep15["Power"], color = "seagreen")

ax[1,0].scatter(ad14.Hour, ad14["Power"], color = "orangered")
ax[1,0].scatter(ad15.Hour, ad15["Power"], color = "seagreen")

ax[1,1].scatter(asd14.Hour, asd14["Power"], color = "orangered")
ax[1,1].scatter(asd15.Hour, asd15["Power"], color = "seagreen")

ax[0,0].set_title("Power consumption for Labour Day (1 May)")
ax[0,1].set_title("Power consumption for Republic Day (2 Jun)")
ax[1,0].set_title("Power consumption for Assumption Day (15 Aug)")
ax[1,1].set_title("Power consumption for All Saints Day (1 Nov)")

ax[0,0].legend(["2015", "2016"])
ax[0,1].legend(["2015"])
ax[1,0].legend(["2014", "2015"])
ax[1,1].legend(["2014", "2015"])

for ax in fig.axes:
    #plt.gca()
    plt.sca(ax)
    ax.set_xlabel("Hour")
    ax.set_ylabel("Power")
    loc = plticker.MultipleLocator(base = 4)
    ax.xaxis.set_major_locator(loc)
    plt.xticks(rotation = 90)
    plt.savefig('graphs\combined_2.png', bbox_inches = "tight")

plt.show()

# graph for 2016 is correct with the dips.
```
